```python
### Modelo de Clasificación: predecir is_high_risk con todas las vars

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

# Eliminamos la variable derivada age_group si existe
df_model = df.drop(columns=["age_group"], errors="ignore")

# Target
y = df_model["is_high_risk"]
X = df_model.drop(columns=["is_high_risk"])

# Columnas categóricas y numéricas
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(include=["int64","float64"]).columns

# Preprocesado
preprocess = ColumnTransformer([
    ("onehot", OneHotEncoder(handle_unknown="ignore"), cat_cols)
], remainder="passthrough")

X_encoded = preprocess.fit_transform(X)

# Train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Modelo
clf = LogisticRegression(max_iter=300)
clf.fit(X_train, y_train)

# Métricas
y_pred = clf.predict(X_test)

print("### CLASSIFICATION REPORT ###")
print(classification_report(y_test, y_pred))

print("### CONFUSION MATRIX ###")
print(confusion_matrix(y_test, y_pred))


feature_names = preprocess.get_feature_names_out()

coef = clf.coef_[0]

coef_df = pd.DataFrame({
    "feature": feature_names,
    "coef": coef
}).sort_values(by="coef", key=abs, ascending=False)

print("\n### TOP 20 FEATURES MÁS INFLUYENTES ###")
print(coef_df.head(20))
```

```python
### CLASIFICACION REDUCIDO:

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import matplotlib.pyplot as plt

## variables relevantes
selected_features = [
    "age",
    "bmi",
    "chronic_count",
    "systolic_bp",
    "diastolic_bp",
    "ldl",
    "risk_score",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "medication_count",
    "smoker"
]

df_model = df[selected_features + ["is_high_risk"]].copy()

### Target
y = df_model["is_high_risk"]

### Predictoras
X = df_model.drop(columns=["is_high_risk"])

### Columnas categóricas y numéricas
cat_cols = ["smoker"]
num_cols = [c for c in selected_features if c not in cat_cols]

### Preprocesado
preprocess = ColumnTransformer([
    ("onehot", OneHotEncoder(drop="first"), cat_cols)
], remainder="passthrough")

X_encoded = preprocess.fit_transform(X)

### Train/Test
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

### Modelo
clf = LogisticRegression(max_iter=300)
clf.fit(X_train, y_train)

### Predicciones
y_pred = clf.predict(X_test)

### Métricas
print("### CLASSIFICATION REPORT ###")
print(classification_report(y_test, y_pred))
print("### CONFUSION MATRIX ###")
print(confusion_matrix(y_test, y_pred))


feature_names = preprocess.get_feature_names_out()

### Coeficientes ordenados por valor absoluto
coef_df = pd.DataFrame({
    "feature": feature_names,
    "coef": clf.coef_[0]
})

coef_df = coef_df.reindex(coef_df["coef"].abs().sort_values(ascending=False).index)

print("\n### COEFICIENTES ORDENADOS POR VALOR ABSOLUTO ###")
print(coef_df)

## grafica ordenada

plt.figure(figsize=(10,6))
plt.barh(coef_df["feature"], coef_df["coef"], color="steelblue")
plt.axvline(x=0, color="black", linewidth=1)
plt.xlabel("Coeficiente")
plt.title("Coeficientes del Modelo de Clasificación (ordenados por |coef|)")
plt.gca().invert_yaxis()  # para mostrar los más importantes arriba
plt.show()
```

```python
### Modelo de Regresión Lineal para annual_medical_cost

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

### vars relevantes

reg_features = [
    "age",
    "bmi",
    "chronic_count",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "medication_count",
    "risk_score",
    "proc_surgery_count",
    "proc_imaging_count",
    "claims_count",
    "avg_claim_amount",
    "total_claims_paid",
    "smoker"
]

df_reg = df[reg_features + ["annual_medical_cost"]].copy()

### Target
y_reg = df_reg["annual_medical_cost"]

### Predictoras
X_reg = df_reg.drop(columns=["annual_medical_cost"])

### Columnas categóricas y numéricas
cat_cols = ["smoker"]
num_cols = [c for c in reg_features if c not in cat_cols]

### Preprocesado
preprocess_reg = ColumnTransformer([
    ("onehot", OneHotEncoder(drop="first"), cat_cols)
], remainder="passthrough")

X_reg_encoded = preprocess_reg.fit_transform(X_reg)

### Train/Test
Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg_encoded, y_reg, test_size=0.2, random_state=42
)

### Modelo
reg = LinearRegression()
reg.fit(Xr_train, yr_train)

### Predicción
yr_pred = reg.predict(Xr_test)

### Métricas
mae = mean_absolute_error(yr_test, yr_pred)
rmse = np.sqrt(mean_squared_error(yr_test, yr_pred))
r2 = r2_score(yr_test, yr_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)

### Nombres de características
reg_feature_names = preprocess_reg.get_feature_names_out()

### Coeficientes ordenados
reg_coef_df = pd.DataFrame({
    "feature": reg_feature_names,
    "coef": reg.coef_
}).sort_values(by="coef", key=abs, ascending=False)

print("\n### TOP 20 COEFICIENTES MÁS INFLUYENTES ###")
print(reg_coef_df.head(20))
```

```python
### Modelo de Regresión : RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
import pandas as pd

### Variables relevantes (mismas que antes)

reg_features = [
    "age",
    "bmi",
    "chronic_count",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "medication_count",
    "risk_score",
    "proc_surgery_count",
    "proc_imaging_count",
    "claims_count",
    "avg_claim_amount",
    "total_claims_paid",
    "smoker"
]

df_rf = df[reg_features + ["annual_medical_cost"]].copy()

### Target
y_rf = df_rf["annual_medical_cost"]

### Predictors
X_rf = df_rf.drop(columns=["annual_medical_cost"])

### Categóricas
cat_cols = ["smoker"]
num_cols = [c for c in X_rf.columns if c not in cat_cols]

### Preprocesado
preprocess_rf = ColumnTransformer([
    ("onehot", OneHotEncoder(drop="first"), cat_cols)
], remainder="passthrough")

X_rf_encoded = preprocess_rf.fit_transform(X_rf)

### Train/Test
Xtr, Xts, ytr, yts = train_test_split(
    X_rf_encoded, y_rf, test_size=0.2, random_state=42
)

### Random Forest
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)

rf.fit(Xtr, ytr)

### Predicción
ypred_rf = rf.predict(Xts)

### Métricas
mae_rf = mean_absolute_error(yts, ypred_rf)
rmse_rf = np.sqrt(mean_squared_error(yts, ypred_rf))
r2_rf = r2_score(yts, ypred_rf)

print("MAE:", mae_rf)
print("RMSE:", rmse_rf)
print("R2 Score:", r2_rf)

### Importancias de variables
importances = rf.feature_importances_
feat_names = preprocess_rf.get_feature_names_out()

rf_importances = pd.DataFrame({
    "feature": feat_names,
    "importance": importances
}).sort_values(by="importance", ascending=False)

print("\n### FEATURE IMPORTANCES ###")
print(rf_importances.head(20))
```

```python
### DASHBOARD

from jupyter_dash import JupyterDash
from dash import dcc, html
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import os


os.system("fuser -k 8052/tcp")

# Crear app con nombre único
app = JupyterDash("medical_dashboard")



# Crear grupos de edad
df["age_group"] = pd.cut(df["age"], 
                         bins=[0,20,30,40,60,80,100,120],
                         labels=["0-20","21-30","31-40","41-60","61-80","81-100","100+"])


#Figura 1: Coste medio por grupo de edad y sexo
mean_age_sex = df.groupby(["age_group", "sex"])["annual_medical_cost"].mean().reset_index()
fig1 = px.bar(mean_age_sex,
              x="age_group",
              y="annual_medical_cost",
              color="sex",
              barmode="group",
              title="Coste medio anual por grupo de edad y sexo")


#Figura 2: Coste medio según hábito de fumar
mean_smoker = df.groupby("smoker")["annual_medical_cost"].mean().reset_index()
fig2 = px.bar(mean_smoker,
              x="smoker",
              y="annual_medical_cost",
              title="Coste medio anual según hábito de fumar")


#Figura 3: Coste medio según número de enfermedades crónicas
mean_chronic = df.groupby("chronic_count")["annual_medical_cost"].mean().reset_index()
fig3 = px.bar(mean_chronic,
              x="chronic_count",
              y="annual_medical_cost",
              title="Coste medio anual según número de enfermedades crónicas")


##coeficientes ordenados


def clean_name(name):
    name = name.replace("remainder__", "")
    name = name.replace("onehot__", "")
    name = name.replace("smoker_", "smk_")
    name = name.replace("hospitalizations_last_3yrs", "hosp_3yrs")
    name = name.replace("days_hospitalized_last_3yrs", "days_hosp_3yrs")
    name = name.replace("visits_last_year", "visits_yr")
    name = name.replace("medication_count", "meds")
    return name

coef_df["clean_feature"] = coef_df["feature"].apply(clean_name)


coef_sorted = coef_df.reindex(coef_df["coef"].abs().sort_values(ascending=False).index)


fig4 = go.Figure()
fig4.add_trace(go.Bar(
    x=coef_sorted["coef"],
    y=coef_sorted["clean_feature"],
    orientation="h",
    marker_color="steelblue"
))

fig4.update_layout(
    title="Coeficientes del Modelo de Clasificación (ordenados por |coef|)",
    xaxis=dict(title="Coeficiente", zeroline=True, zerolinewidth=2, zerolinecolor="black"),
    yaxis=dict(title="Variable"),
    height=600,
    margin=dict(l=150, r=20, t=50, b=40),
    showlegend=False
)


# ============================
# LAYOUT
# ============================

app.layout = html.Div([

    html.H1("Dashboard de Análisis de Costes Médicos", style={"textAlign": "center"}),

    html.Div([
        html.Div([dcc.Graph(figure=fig1)], style={"width": "48%", "display": "inline-block", "padding": "1%"}),
        html.Div([dcc.Graph(figure=fig2)], style={"width": "48%", "display": "inline-block", "padding": "1%"})
    ]),

    html.Div([
        html.Div([dcc.Graph(figure=fig3)], style={"width": "48%", "display": "inline-block", "padding": "1%"}),
        html.Div([dcc.Graph(figure=fig4)], style={"width": "48%", "display": "inline-block", "padding": "1%"})
    ]),
])

# ============================
# EJECUCIÓN
# ============================

app.run_server(mode="inline", port=8057)
```
